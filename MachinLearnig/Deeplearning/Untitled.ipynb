```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("./DACONLG/train.csv")
train.head()
```

```python
train = train.drop(["ID", "X_03", "X_23", "X_47", "X_48"], axis=1)
```

```python
train_1 = train.iloc[:14771, :]
train_1
```

```python
train_2 = train.iloc[14771:, :]
train_2
```

```python
X_train_1 = train_1.iloc[:, :52]
y_train_1 = train_1.iloc[:, 52:]

X_train_2 = train_2.iloc[:, :52]
y_train_2 = train_2.iloc[:, 52:]

X_train_1.head()
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train_1_scaled = ss.fit_transform(X_train_1)
X_train_2_scaled = ss.fit_transform(X_train_2)
```

```python
y_train_1.head()
```

```python
X_test = pd.read_csv("./DACONLG/test.csv")
X_test.head()
```

```python
X_test = X_test.drop(["ID", "X_03", "X_23", "X_47", "X_48"], axis=1)
```

```python
X_test_1 = X_test.iloc[:14771, :]
X_test_2 = X_test.iloc[14771:, :]
```

```python
X_test_1_scaled = ss.fit_transform(X_test_1)
X_test_2_scaled = ss.fit_transform(X_test_2)
```

```python
from sklearn.multioutput import MultiOutputRegressor
from xgboost import XGBRegressor
```

```python
xg1 = MultiOutputRegressor(XGBRegressor(n_estimators=60, learning_rate=0.03, gamma = 0, subsample=1, colsample_bytree = 1, max_depth=9)).fit(X_train_1_scaled, y_train_1)
xg2 = MultiOutputRegressor(XGBRegressor(n_estimators=60, learning_rate=0.03, gamma = 0, subsample=1, colsample_bytree = 1, max_depth=9)).fit(X_train_2_scaled, y_train_2)
```

```python
y_pred_1 = xg1.predict(X_test_1_scaled)
y_pred_2 = xg2.predict(X_test_2_scaled)
```

```python
y_pred_1.shape, y_pred_2.shape
```

```python
y_pred_1
```

```python
y_pred_2
```

```python
y_pred = np.concatenate((y_pred_1, y_pred_2), axis=0)
y_pred
```

```python
len(y_pred)
```

```python
submit = pd.read_csv('./DACONLG/sample_submission.csv')
```

```python
for idx, col in enumerate(submit.columns):
    if col=='ID':
        continue
    submit[col] = y_pred[:,idx-1]
print('Done.')
```

```python
submit.to_csv('./DACONLG/submit.csv', index=False)
```
