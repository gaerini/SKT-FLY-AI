# 결측치 처리

```python
df.head()
```

```python
df = df.drop(["PassengerId", "Cabin"], axis=1)
df.head()
```

```python
df = df.drop("Ticket", axis = 1)
df.head()
```

```python
#Survive : 생존 1, 사망 0
#Pclass : 1st:0, 2nd:1, 3rd:2
# Sex : male:1, female:0
# Embarked : c q s -> 0, 1, 2
```

```python
def f1(x):
    if x == 'lost':
        return 0
    else:
        return 1
```

```python
df["Survived"] = df["Survived"].apply(f1)
df.head()
```

```python
def f2(x):
    if x == '1st':
        return 0
    elif x == '2nd':
        return 1
    else:
        return 2
```

```python
df["Pclass"] = df["Pclass"].apply(f2)
df.head()
```

```python
def f3(x):
    if x == 'male':
        return 1
    else:
        return 0
```

```python
df["Sex"] = df["Sex"].apply(f3)
df.head()
```

```python
def f4(x):
    if x == 'C':
        return 0
    elif x == 'Q':
        return 1
    else:
        return 2
```

```python
df["Embarked"] = df["Embarked"].apply(f4)
df.head()
```

```python
df.isna().sum()
```

```python
df.head()
```

## Age 결측치 해결

```python
# 나이가 NaN 인 데이터 표시

df.isna().sum()
```

```python
df[df['Age'].isna()]
```

```python
df['Initial']=df.Name.str.extract("([A-Za-z]+)\.")
```

```python
df.head(10)
```

```python
df['Initial'].unique()
```

```python
be = ['Mlle', 'Mme', 'Ms', 'Dr', 'Major', 'Lady', 'Countess', 'Johnkeer', 'Col', 'Rev', 'Capt', 'Sir', 'Don']
ae = ['Miss', 'Mrs', 'Miss', 'Mr', 'Mr', 'Mrs', 'Mrs', 'Other', 'Other', 'Other', 'Mr', 'Mr', 'Mr']
```

```python
df['Initial']=df['Initial'].replace(be, ae)
df['Initial'].unique()
```

```python
#호칭별 나이 평균
am = round(df.groupby("Initial").Age.mean(),1)
am
```

```python
df.loc[(df["Initial"]=="Mr") & df["Age"].isna(), "Age"] = am["Mr"]
```

```python
df.loc[(df["Initial"]=="Jonkheer") & df["Age"].isna(), "Age"] = am["Jonkheer"]
df.loc[(df["Initial"]=="Master") & df["Age"].isna(), "Age"] = am["Master"]
df.loc[(df["Initial"]=="Miss") & df["Age"].isna(), "Age"] = am["Miss"]
df.loc[(df["Initial"]=="Mrs") & df["Age"].isna(), "Age"] = am["Mrs"]
df.loc[(df["Initial"]=="Other") & df["Age"].isna(), "Age"] = am["Other"]
```

```python
df.isna().sum(axis=0)
```

```python
# 항구 onehot encoding
x = pd.get_dummies(df["Embarked"])
x.columns=['C', 'Q', 'S']
x
```

```python
df = pd.concat([df, x], axis=1)
df.head()
```

```python
df = df.drop(['Embarked','Initial'], axis=1)
df.head()
```

```python
df = df.drop('Name', axis=1)
df.head()
```
